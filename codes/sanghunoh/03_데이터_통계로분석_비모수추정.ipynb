#비모수 확률밀도 추정(NonParametric Density Estimation)


*   데이터로부터 직접 밀도 추정




## 히스토그램법
$
* i번째 구간 확률값 \\
Pr[x_0+(i-1]h <= x < x_0+ih)] = \frac{K(x \in [x_0+(i-1)h, x_0+ih))}{N}  (분모: 해당 구간 속한 데이터 개수, 분모 : 전체 데이터 개수) \\
* n차원 데이터 x에 대한 확률밀도 함수 추정식 \\
p(x) = \frac{1}{h^n}\frac{K(x)}{N}  (h^n : 구간 부피)
$
*   알맞은 h값 선정이 추정에 큰 영향
*   매끄럽지 못한 추정 분포




```python
import numpy as np
row_count = 100
x_rand = np.sort(np.random.randn(row_count))  # 그래프 분포 점 표시 위해 정렬
x_rand.shape, x_rand.mean(), x_rand.std()# x_rand
```

## 파라미터 h값에 따른 추정 변화

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot(x_rand, bins=5, kde=True, rug=True, rug_kws={'height':0.1})
sns.displot(x_rand, bins=10, kde=True, rug=True, rug_kws={'height':0.1})
sns.displot(x_rand, bins=20, kde=True, rug=True, rug_kws={'height':0.1})
plt.show()
```
