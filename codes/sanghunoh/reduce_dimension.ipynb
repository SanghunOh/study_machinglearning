```python
import numpy as np
row_count = 20
dimension = 10
```

```python
rand_std_1 = np.random.randn(row_count, dimension)
# rand_std_1
rand_std_1.shape # 행 * 열(데이터를 나타내는 성격)
```

```python
# identity vector
identity_vector = np.identity(n=dimension, dtype=np.int8)
identity_vector
```

```python
# projection 3 dimension
projection_vector = identity_vector[:,[1,4,9]]
rand_std_1.shape, projection_vector.shape
```

```python
# row(row_count) * dimension(number reduction)
projected_vector = np.matmul(rand_std_1, projection_vector)
projected_vector
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(14,5))
ax = fig.add_subplot(121, projection='3d')
ax.scatter(projected_vector[:,0],projected_vector[:,1],projected_vector[:,2])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
